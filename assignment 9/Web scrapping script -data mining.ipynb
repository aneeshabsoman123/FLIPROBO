```python
Price=[]
Brand=[]
distancegas=[]
distance=[]
gas=[]
Model=[]
```

```python
for i in range(500):
            driver.execute_script("window.scrollBy(0, 250)")
            time.sleep(3)
```

```python
price=driver.find_elements_by_xpath("//div[@class='_7udZZ']//span")
for i in price:
    try:
        Price.append(i.text)
    except NoSuchElementException:
        Price.append("--")
        
```

```python



brand=driver.find_elements_by_xpath("//h2[@class='_3FpCg']")
for i in brand:
    try:
        Brand.append(i.text)
    except NoSuchElementException:
        Brand.append("--")
        

Distancegas=driver.find_elements_by_xpath("//ul[@class='bVR0c']")
for i in Distancegas:
    try:
        distancegas.append(i.text)
    except NoSuchElementException:
        distancegas.append("--")
```

```python
#creating dataframe
df=pd.DataFrame({'Price':Price,
                'Brand':Brand,
                'distancegas':distancegas})
df
```

```python
import re

Brandsplit=[]
for i in range(2289):
        s = df["Brand"][i]
        res = re.split('(\d+)', s)
        Brandsplit.append(res)
        
```

```python
dataframe=pd.DataFrame(Brandsplit, columns=['a','b','c','d','e','f','g'])
```

```python
dataframe
```

```python
df['year']=dataframe['b']
df['BrandModel']=dataframe['c']  
df.drop(columns=["Brand"],inplace=True)
df
```

```python
#splitting the data
df[['distance', 'owner', 'Gas']] = df['distancegas'].str.split('\n', 2, expand=True)
df.drop(columns=["distancegas"],inplace=True)

```

```python
df
```

```python
df2 = df['BrandModel'].str.split(n=1, expand=True)
df2.columns = ['STATUS_ID{}'.format(x+1) for x in df2.columns]
df = df.join(df2)

#renaming statusid1 and statusid2
df.rename(columns={'STATUS_ID1': 'Brand',
                   'STATUS_ID2': 'Model'},
          inplace=True, errors='raise')
```
