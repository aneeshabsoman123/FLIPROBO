# 1. Write a python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’.

```python
#importing required libraries
from bs4 import BeautifulSoup
import requests

#source code is stored in page
page1=requests.get("https://en.wikipedia.org/wiki/Main_Page")
page1
```

```python
#creating soup and header names
soup1=BeautifulSoup(page1.content)

heading_tags = ["h1", "h2", "h3"]
for tags in soup1.find_all(heading_tags):
    print(tags.name + ' -> ' + tags.text.strip())
```

# 2.python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and make data frame

```python
#importing required libraries
from bs4 import BeautifulSoup
import requests

#source code is stored in page
page2=requests.get("http://www.imdb.com/chart/top")
page2

soup2=BeautifulSoup(page2.content)
```

```python
#retrieving movies data
#method1
movies= soup2.findAll('td', {'class': 'titleColumn'})
movies

#method2
movies=soup2.select('td.titleColumn')
movies

#iterating to the list and appending it
movie_titles=[]
for i in movies:
    movie_titles.append(i.text.strip("\n"))
movie_titles
```

```python
year = soup2.findAll("span",{"class":"secondaryInfo"})
year

#iterating to the list and appending it
year_titles=[]
for i in year:
    year_titles.append(i.text.strip())
year_titles
```

```python

#rating
rating= soup2.findAll('td', {'class': 'ratingColumn imdbRating'})
rating

#iterating to the list and appending it
rating_titles=[]
for i in rating:
    rating_titles.append(i.text.strip('\n'))
rating_titles
```

```python
import pandas as pd
imbd=pd.DataFrame({})
imbd['title']=movie_titles
imbd['year']=year_titles
imbd['rating']=rating_titles
imbd
```

# 4. program to scrap book name, author name, genre and book review of any 5 books from
‘www.bookpage.com

```python
#importing required libraries
from bs4 import BeautifulSoup
import requests

#source code is stored in page
page3=requests.get("https://bookpage.com/")

soup3=BeautifulSoup(page3.content)

    
```

```python
#book name
book_name= soup3.findAll('h4', {'class': 'italic'})
book_name

#iterating to the list and appending it
book_titles=[]
for i in book_name:
    book_titles.append(i.text.strip('\n'))
book_titles
```

```python
#author name
author_name=soup3.findAll('p', {'class': 'sans bold'})
author_name

#iterating to the list and appending it
author_titles=[]
for i in author_name:
    author_titles.append(i.text.strip(','))
author_titles
```

```python
#genre

genre=soup3.findAll('p', {'class': 'genre-links hidden-phone'})


#iterating to the list and appending it
genre_titles=[]
for i in genre:
    genre_titles.append(i.text.strip('\n/ \n'))
genre_titles
```

```python
#book review

book_review=soup3.findAll('p', {'class': 'excerpt'})


#iterating to the list and appending it
bookreview_titles=[]
for i in book_review:
    bookreview_titles.append(i.text.strip(','))
bookreview_titles
```

```python
import pandas as pd
bookpage=pd.DataFrame({})
bookpage['title']=book_titles
bookpage['author']=author_titles
bookpage['genre']=genre_titles
bookpage
```

# 5. python program to scrape cricket rankings from ‘www.icc-cricket.co

a.Top 10 ODI teams in men’s cricket along with the records for matches, points and rating

```python
page5a=requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/test")
soup5a=BeautifulSoup(page5a.content)
page5a


table1= soup5a.findAll("table", { "class" : "table " })
table1
 
rows1=[]
#find all tr tag(rows)-traversing rows 
for row in table1:
    rows1 = table1.find_all('tr')

#find all td tags(columns)-traversing columns inside the rows
for tr1 in rows1:
    data1=[]
    cols1 = tr1.find_all('td') 
    for td1 in cols1:
        data1.append(td1.text.strip()) 
    print(data1)
```

# 7.python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in.

```python
# mobile Names titles
mobile=soup6.findAll('span', {'class': 'a-size-medium a-color-base a-text-normal'})
mobile

#iterating to the list and appending it
mobile_titles=[]
for i in mobile:
    mobile_titles.append(i.text.strip(','))
mobile_titles
```

```python
#Price

mobilePrice=soup6.findAll('span', {'class': 'a-price-whole'})


mobilePrice_titles=[]
for i in mobilePrice:
    mobilePrice_titles.append(i.text.strip(','))
mobilePrice_titles
```

```python
# Image URL

Image_URL=soup6.find_all('a',{'class': 'a-link-normal'})
 
URL_titles=[]
for i in Image_URL:
    URL_titles.append(i.text)
URL_titles
```

```python
#Average Rating
avgRating=soup6.findAll('span', {'class': 'a-icon-alt'})

avgRating_titles=[]
for i in avgRating:                              
    avgRating_titles.append(i.text.strip(','))
    
avgRating_titles
```

# 9.scrape fresher job listings from ‘https://internshala.com/’. It should include job title,

```python
import pandas as pd
jobs=pd.DataFrame({})
jobs['title']=job_titles
jobs['company']=company_titles
jobs['stipend']=stipend_titles

jobs
```
